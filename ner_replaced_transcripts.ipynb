```python
import json
import pprint
import pandas as pd
```

```python
f = open("ner_replaced_transcripts.txt", "w")
f.write("")
f.close()
```

```python
with open('ner_output_v1_test.json') as f:
    json_data = json.load(f)

pprint.pp(json_data)
```

```python
for each in json_data:
    replaced_text = each.get("Replaced Sentence")
    f = open("ner_replaced_transcripts.txt", "a")
    f.write(replaced_text)
    f.write("\n")
    f.close()
```

```python
with open('ner_output_v1_train.json') as f:
    json_data = json.load(f)

pprint.pp(json_data)
for each in json_data:
    replaced_text = each.get("Replaced Sentence")
    if replaced_text != None: 
        f = open("ner_replaced_transcripts.txt", "a")
        f.write(replaced_text)
        f.write("\n")
        f.close()
```
